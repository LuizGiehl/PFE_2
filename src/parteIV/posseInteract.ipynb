```python
import pandas as pd 

posse = pd.read_csv('../../data/Formulário sem título.csv')
```

```python
print(posse)
print(posse.info())
```

```python
posse['Quantas pessoas vão ?'] = pd.to_numeric(posse['Quantas pessoas vão ?'], errors='coerce')
print(posse)
```

```python
print(posse[['Nome do interactiano',
       'Quantas pessoas vão ?']])
```

```python
posse.loc[1, 'Quantas pessoas vão ?'] = 1
posse.loc[13, 'Quantas pessoas vão ?'] = 2
posse.loc[17, 'Quantas pessoas vão ?'] = 1
posse.loc[19, 'Quantas pessoas vão ?'] = 1
```

```python
print(posse[['Nome do interactiano',
       'Quantas pessoas vão ?']])

posse = posse.rename(columns={'Nome do interactiano': 'Interactiano',
                      'Quantas pessoas vão ?': 'Quantidade_de_pessoas'})

posse = posse.drop(columns=['Carimbo de data/hora'])
posse = posse.drop(index=14)
```

```python
total = pd.DataFrame({'Interactiano': 'Total de pessoas', 'Quantidade_de_pessoas': [posse['Quantidade_de_pessoas'].sum()]})
posse = pd.concat([posse, total], ignore_index=True)
```

```python
posse_table = posse.to_html()

from IPython.display import HTML

HTML(posse_table)
```
